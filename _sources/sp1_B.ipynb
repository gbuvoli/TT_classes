## Mínimo, Máximo, Sumar y Ordenar elementos

`min()`, `max()`, `sum()`, `sort()`

### 📊 Calcular el Total de una Compra (sum())
Dada una lista de precios de productos en una compra, calcula el total:

```python
precios = [10.50, 25.30, 5.99, 12.40, 7.80]
total = sum(precios)

print(f"Total a pagar: ${total}")  # Ese valor tiene muchos decimales

print(f"Total a pagar: ${round(total,2)}")  #Uso de round() para redondear
```

###💰 Encontrar el Producto Más Caro y Más Barato (max() y min())

```python
precios = [10.50, 25.30, 5.99, 12.40, 7.80]
mas_caro = max(precios)
mas_barato = min(precios)

print(f"Producto más caro: ${round(mas_caro,2)}")
print(f"Producto más barato: ${round(mas_barato,2)}")

promedio= sum(precios)/len(precios)
print("El valor promedio es: ", promedio, " esto es la continuacion")
```

###📦 Ordenar Productos por Precio (sort())

```python
precios = [10.50, 25.30, 5.99, 12.40, 7.80]
precios.sort()  # Ordenar de menor a mayor
print("Precios ordenados (barato a caro):", precios)

precios.sort(reverse=True)  # Ordenar de mayor a menor
print("Precios ordenados (caro a barato):", precios)
```

### 🛒 .sort() vs. sorted()

* `sort()` modifica la lista original
* `sorted()` devuelve una nueva lista ordenada sin cambiar la original.

```python

# En esta ocasion, exploremos una lista de tipo String.

productos = ["pollo", "res", "cerdo", "pescado","pavo"]
print("Lista antes de ordenar:", productos)

# Usando sort() (modifica la lista)
productos.sort()
print("Lista ordenada con sort():", productos)  # La lista original cambia

productos.sort(reverse=True) #Utilizamos reverse para invertir el orden, ahora es descendente
print("Lista ordenada en orden inverso con sort():", productos)
```

```python
# Usando sorted() (no modifica la lista original)
precios_original = [15, 5, 20, 10, 25]
print(f"Usamos sorted() sobre la lista original: {sorted(precios_original)}")
print(f"Ahora revisemos la lista original a ver si cambió: {precios_original}")
print("La lista original no se modificó")
```



> Para conservar los datos resultantes del método `sorted()` se deben guardar en una variable. Ya que no modifica la lista original.



```python
#Probemos ahora:
precios_nueva = sorted(precios_original)

print("Lista original después de sorted():", precios_original)  # Se mantiene igual
print("Lista ordenada con sorted():", precios_nueva)  # Nueva lista ordenada
```

 ## 📌 Explicación rápida de slicing:


* `lista[inicio:fin]` → Extrae elementos desde inicio hasta **fin-1**.
* `lista[:n]` → Desde el inicio hasta **n-1**.
* `lista[-n:]` → Últimos n elementos.

### 🚀 LLevalo a otro nivel!

* `lista[::salto]` → Salta salto elementos.
* `lista[::-1]` → Invierte la lista.



> 👉 **PRO TIP:**
- Recuerda que en la indexación **positiva** iniciamos a contar desde `0`.
- En la indexación **Negativa**, comenzamos desde el `-1`
- La indexación `-0` **No existe**


```python
# Obtener una sublista que incluya los primeros 4 productos.

productos = ["Arroz", "Frijoles", "Leche", "Pan", "Huevos", "Queso", "Jugo", "Carne", "Pasta", "Manzana"]

primeros_cuatro = productos[0:4]
primeros_cuatro_otro = productos[:4]

print(primeros_cuatro)
print(primeros_cuatro_otro)
```

```python
# Obtener una sublista desde el tercer producto hasta el séptimo usando indexacion positiva.

productos = ["Arroz", "Frijoles", "Leche", "Pan", "Huevos", "Queso", "Jugo", "Carne", "Pasta", "Manzana"]

sublista = productos[2:7] # Recuerda que las posiciones se inician a contar en 0 y se escriben hasta la posición final-1

print(sublista)
```

```python
# Obtener una sublista desde el tercer producto hasta el séptimo pero utilizando indexación negativa.

productos = ["Arroz", "Frijoles", "Leche", "Pan", "Huevos", "Queso", "Jugo", "Carne", "Pasta", "Manzana"]

sublista_positiva = productos[2:7]
sublista_negativa = productos[-8:]  # Recuerda que las posiciones se inician a contar en 0 y se escriben hasta posicion final -1

print(sublista_positiva)
print(sublista_negativa)
```

```python
# Acceder a los ultimos registros

#Queremos los ultimos 3 productos de la lista.

productos = ["Arroz", "Frijoles", "Leche", "Pan", "Huevos", "Queso", "Jugo", "Carne", "Pasta", "Manzana"]

print(productos[-3:0])
print(productos[-3:-1])

print(productos[-3:], " Esta es la forma correcta de hacerlo")
```

# 5. Modificación de listas: .append(), .extend(), .pop(), .insert(), .remove()
---

Supongamos que estamos generando una factura, pero el cliente realiza algunos cambios en su compra:

📌 **Situación inicial:**
- El cliente ha seleccionado algunos productos, pero decide agregar más, eliminar otros y actualizar una promoción.

```python
# Lista inicial con productos en la factura (extraídos de la lista de supermercado)

factura = [
    ["Manzanas", 2.5, "kg", False, 2],  # Producto, precio, unidad, promo, cantidad
    ["Leche", 3.0, "litro", False, 1],
    ["Pollo", 5.5, "kg", True, 1],
]

# 1️⃣ El cliente decide agregar un nuevo producto a la factura
factura.append(["Galletas", 2.5, "paquete", True, 2])  # Se agregan galletas en promo

# 2️⃣ El cliente cambia de opinión y elimina el pollo de la factura

factura.pop(2)  # Se elimina el tercer elemento (índice 2)

# 3️⃣ El cliente quiere añadir una combo especial que está en oferta, lleva 1kg de queso y te regalamos 1 litro de jugo de naranja.
factura.extend([
    ["Queso", 4.8, "kg", False, 1],
    ["Jugo de naranja", 0, "litro", True, 1]
])

# 4️⃣ Se da cuenta de que olvidó incluir papel higiénico, lo agrega en la segunda posición
factura.insert(1, ["Papel Higiénico", 5.0, "pack", False, 1])

# 5️⃣ El supermercado ha cambiado la promoción: ahora las manzanas están en oferta
factura[0][3] = True  # Se actualiza la promoción de las manzanas

# 6️⃣ El cliente decide eliminar un producto que ya no quiere
factura.remove(["Queso", 4.8, "kg", False, 1])  # Se elimina el queso

# 📌 **Imprimir la factura final**
print("\nFactura de venta #0001 - Supermercado TripleTEN\n")
print("Producto  Cantidad  Precio Subtotal")
print("----------------------------------")
print(f"{factura[0][0]} - {factura[0][4]} x ${factura[0][1]} = ${factura[0][1] * factura[0][4]}")
print(f"{factura[1][0]} - {factura[1][4]} x ${factura[1][1]} = ${factura[1][1] * factura[1][4]}")
print(f"{factura[2][0]} - {factura[2][4]} x ${factura[2][1]} = ${factura[2][1] * factura[2][4]}")
print(f"{factura[3][0]} - {factura[3][4]} x ${factura[3][1]} = ${factura[3][1] * factura[3][4]}")
print(f"{factura[4][0]} - {factura[4][4]} x ${factura[4][1]} = ${factura[4][1] * factura[4][4]}")
print("----------------------------------")
```

## 📝 Resumen de lo aprendido sobre listas y listas anidadas

1. 📌 Conceptos clave sobre listas:

  * Son estructuras de datos mutables que permiten almacenar múltiples elementos.
  * Pueden contener distintos tipos de datos (números, cadenas, booleanos e incluso otras listas).
  * Se accede a sus elementos mediante índices.
  * Se pueden modificar, agregar o eliminar elementos con métodos específicos.

2. 📌 Listas anidadas (listas dentro de listas)
  * Permiten representar estructuras complejas como tablas o matrices.
  * Se accede a sus elementos mediante índices múltiples.
  * Útiles para organizar datos jerárquicos como un supermercado con categorías de productos.
3. 📌 Métodos importantes de listas:

  * append: Agrega un elemento al final de la lista.
  * extend: Agrega múltiples elementos a la lista.
  * insert: Inserta un elemento en una posición específica.
  * pop: Elimina y devuelve un elemento de una posición dada.
  * remove: Elimina la primera aparición de un elemento.
  * sort: Ordena la lista en orden ascendente.
  * reverse: Invierte el orden de la lista.

4. 📌 Indexación y slicing (rebanado):
  * Se puede acceder a elementos individuales mediante índices positivos y negativos.
  * Se pueden extraer porciones de la lista utilizando rangos de índices.

5. 📌 Aplicación en facturación del supermercado
  * Uso de listas anidadas para organizar productos por categoría.
  * Cálculo de subtotales a partir de las listas.
  * Modificación de precios y productos dentro de la lista.
  * Eliminación de productos discontinuados.
  * Inserción de nuevos productos en la lista según promociones o stock.
